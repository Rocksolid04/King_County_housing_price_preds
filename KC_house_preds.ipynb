```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=256,epochs=400)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
preds=model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,preds))
```

```python
mean_absolute_error(y_test,preds)
```

```python
# Define feature columns (exclude 'price')
feature_cols = [col for col in df.columns if col != "price"]

# X = only features
X = df[feature_cols]
y = df["price"]

# Median defaults (19 values only)
default_values = X.median().values

def predict_price_with_5(model, scaler, bedrooms, bathrooms, sqft_living, waterfront, condition, default_values, feature_cols):
    # Start with defaults
    features = default_values.copy()

    # Replace chosen features by index
    features[feature_cols.index("bedrooms")] = bedrooms
    features[feature_cols.index("bathrooms")] = bathrooms
    features[feature_cols.index("sqft_living")] = sqft_living
    features[feature_cols.index("waterfront")] = waterfront
    features[feature_cols.index("condition")] = condition

    # Scale and predict
    features_scaled = scaler.transform([features])
    prediction = model.predict(features_scaled)
    return float(prediction[0])
```

```python
### Variable
bedroom = 3
bathroom = 2
sq_ft = 1000
condition = 5
waterfront = 1
```

```python
print(predict_price_with_5(model,scaler,bedroom,bathroom,sq_ft,waterfront,condition,default_values,feature_cols))
```
